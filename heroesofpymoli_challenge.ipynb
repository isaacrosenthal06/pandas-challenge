```python
import pandas as pd
purchase_file = '../Resources/purchase_data.csv'

purchase_df = pd.read_csv(purchase_file)
purchase_df
```

```python
totalplayers = len(purchase_df['SN'].unique())
totalplayers
players_df = pd.DataFrame(totalplayers, index = [0], columns = ['Total Players'])
#players_df['Total Players'] = totalplayers
players_df
```

```python
summary_df = purchase_df
summary_df

uniqueitems = len(summary_df['Item Name'].unique())
averageprice = summary_df['Price'].mean()
numberofpurchases = len(summary_df['Purchase ID'])
totalrevenue = summary_df['Price'].sum()
cleansummary_df = pd.DataFrame(index = [0], columns = ['Number of Unique Items', 'Average Price', 'Number of Purchases', 'Total Revenue'])
cleansummary_df['Number of Unique Items'] = uniqueitems
cleansummary_df['Average Price'] = averageprice
cleansummary_df['Number of Purchases'] = numberofpurchases
cleansummary_df['Total Revenue'] = totalrevenue
cleansummary_df['Total Revenue'] = cleansummary_df['Total Revenue'].apply(lambda x: f"${x}")
cleansummary_df['Average Price'] = round(cleansummary_df['Average Price'], 2).apply(lambda x: f"${x}")


cleansummary_df
```

```python
gender_df = purchase_df
gender_df

totalmales = len(gender_df.loc[(gender_df['Gender'] == 'Male'), 'SN'].unique())
totalmales
totalfemales = len(gender_df.loc[(gender_df['Gender'] == 'Female'), 'SN'].unique())
totalfemales
totalothers = len(gender_df.loc[(gender_df['Gender'] == 'Other / Non-Disclosed'), 'SN'].unique())
totalothers

percentmale = round((totalmales / totalplayers) * 100, 2)
percentfemale = round((totalfemales / totalplayers) * 100, 2)
percentothers = round((totalothers / totalplayers) * 100, 2)

cleangender_df = pd.DataFrame(index = ['Male', 'Female', 'Other / Non-Disclosed' ], columns = ['Total Count', 'Percentage of Players'])
cleangender_df.iloc[0, :] = [totalmales , percentmale]
cleangender_df.iloc[1, :] = [totalfemales, percentfemale]
cleangender_df.iloc[2, :] = [totalothers, percentothers]
cleangender_df['Percentage of Players'] = cleangender_df['Percentage of Players'].apply(lambda x: f"{x}%")

cleangender_df
```

```python
purchasecount = gender_df.loc[: , 'Purchase ID'].groupby(gender_df['Gender']).count()
purchasecount
averagepurchaseprice = round(gender_df.loc[: , 'Price'].groupby(gender_df['Gender']).mean(), 2)
averagepurchaseprice
totalpurchasevalue = round(gender_df.loc[: , 'Price'].groupby(gender_df['Gender']).sum(), 2)
totalpurchasevalue
femaleavgtotalpurch = round(totalpurchasevalue[0] / len(gender_df.loc[(gender_df['Gender'] == 'Female'), 'SN'].unique()), 2)
femaleavgtotalpurch
maleavgtotalpurch = round(totalpurchasevalue[1] / len(gender_df.loc[(gender_df['Gender'] == 'Male'), 'SN'].unique()), 2)
maleavgtotalpurch
otheravgtotalpurch = round(totalpurchasevalue[2] / len(gender_df.loc[(gender_df['Gender'] == 'Other / Non-Disclosed'), 'SN'].unique()), 2)
otheravgtotalpurch

purchase_ana_df = pd.DataFrame(index = ['Male', 'Female', 'Other / Non-Disclosed' ], columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value' , 'Avg Total Purchase per Person'])
purchase_ana_df['Purchase Count'] = purchasecount
purchase_ana_df['Average Purchase Price'] = averagepurchaseprice
purchase_ana_df['Total Purchase Value'] = totalpurchasevalue
purchase_ana_df.iloc[0, 3] = maleavgtotalpurch
purchase_ana_df.iloc[1, 3] = femaleavgtotalpurch
purchase_ana_df.iloc[2, 3] = otheravgtotalpurch
purchase_ana_df['Average Purchase Price'] = purchase_ana_df['Average Purchase Price'].apply(lambda x: f"${x}")
purchase_ana_df['Total Purchase Value'] = purchase_ana_df['Total Purchase Value'].apply(lambda x: f"${x}")
purchase_ana_df['Avg Total Purchase per Person'] = purchase_ana_df['Avg Total Purchase per Person'].apply(lambda x: f"${x}")
purchase_ana_df
```

```python
age_df = purchase_df
age_df
bins = [0, 9, 14, 19, 24, 29, 34, 39, 40]
binlabels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
age_df['Age Group'] = pd.cut(age_df['Age'], bins, labels = binlabels)
groupedage_df = age_df.set_index(age_df['Age Group'])
#groupedage_df = age_df.groupby('Age Group').count()
under10count = len(groupedage_df.loc[(groupedage_df['Age Group'] == '<10'), 'SN'].unique())
under10count
count10to14 = len(groupedage_df.loc[(groupedage_df['Age Group'] == '10-14'), 'SN'].unique())
count10to14
count15to19 = len(groupedage_df.loc[(groupedage_df['Age Group'] == '15-19'), 'SN'].unique())
count15to19
count20to24 = len(groupedage_df.loc[(groupedage_df['Age Group'] == '20-24'), 'SN'].unique())
count20to24
count25to29 = len(groupedage_df.loc[(groupedage_df['Age Group'] == '25-29'), 'SN'].unique())
count25to29
count30to34 = len(groupedage_df.loc[(groupedage_df['Age Group'] == '30-34'), 'SN'].unique())
count30to34
count35to39 = len(groupedage_df.loc[(groupedage_df['Age Group'] == '35-39'), 'SN'].unique())
count35to39
count40plus = len(groupedage_df.loc[(groupedage_df['Age Group'] == '40+'), 'SN'].unique())
count40plus

percentofplayers =[]
countagegroup = [under10count, count10to14, count15to19, count20to24, count25to29, count30to34, count35to39, count40plus]
for number in countagegroup:
    percentofplayers.append(round((number / totalplayers) * 100, 2))
    
percentofplayers

byagegroups_df = pd.DataFrame(countagegroup, index = binlabels, columns= ['Total Count'])
byagegroups_df['Percentage of Players'] = percentofplayers
byagegroups_df['Percentage of Players'] = byagegroups_df['Percentage of Players'].apply(lambda x: f"{x}%")
byagegroups_df
#groupedage_df.head()
```

```python
agegroups_df = purchase_df
agegroups_df['Age Group'] = pd.cut(agegroups_df['Age'], bins, labels = binlabels)
groupedagegroups_df = agegroups_df.set_index(age_df['Age Group'])
groupedagegroups_df = agegroups_df.groupby('Age Group').count()[['Purchase ID']]
groupedagegroups_df['Average Purchase Price'] = agegroups_df.groupby('Age Group').mean()[['Price']]
groupedagegroups_df['Total Purchase Value'] = agegroups_df.groupby('Age Group').sum()[['Price']]

under10avgpurch = round((groupedagegroups_df.iloc[0, 2] / under10count), 2)
under10avgpurch
avgpurch10to14 = round((groupedagegroups_df.iloc[1, 2] / count10to14), 2)
avgpurch10to14
avgpurch15to19 = round((groupedagegroups_df.iloc[2, 2] / count15to19), 2)
avgpurch15to19
avgpurch20to24 = round((groupedagegroups_df.iloc[3, 2] / count20to24), 2)
avgpurch20to24
avgpurch25to29 = round((groupedagegroups_df.iloc[4, 2] / count25to29), 2)
avgpurch25to29
avgpurch30to34 = round((groupedagegroups_df.iloc[5, 2] / count30to34), 2)
avgpurch30to34
avgpurch35to39 = round((groupedagegroups_df.iloc[6, 2] / count35to39), 2)
avgpurch35to39
avgpurch40plus =  round((groupedagegroups_df.iloc[7, 2] / count40plus), 2)
avgpurch40plus
avgpurchagegroups = [under10avgpurch, avgpurch10to14, avgpurch15to19, avgpurch20to24, avgpurch25to29, avgpurch30to34, avgpurch35to39, avgpurch40plus]
groupedagegroups_df['Avg Total Purchase per Person'] = avgpurchagegroups
groupedagegroups_df
```

```python
spender_df = purchase_df.set_index(purchase_df['SN'])
spender_df = purchase_df.groupby('SN').count()[['Purchase ID']]
spender_df['Average Purchase Price'] = purchase_df.groupby('SN').mean()[['Price']]
spender_df['Total Purchase Value'] = purchase_df.groupby('SN').sum()[['Price']]

spender_df = spender_df.sort_values('Purchase ID', ascending= False)

spender_df = spender_df.rename(columns = {'Purchase ID' : 'Purchase Count'})

spender_df.head() 
```

```python
purchase_df
item_df = purchase_df[['Item ID', 'Item Name', 'Price']]
item_df
#groupeditem_df = item_df.set_index(item_df['Item ID'])
groupeditempurchasecount = item_df.groupby(['Item ID', 'Item Name']).count()['Price']
groupeditemavgprice = item_df.groupby(['Item ID', 'Item Name']).mean()['Price']
groupeditemavgprice
groupeditemtotalpurchasevalue = groupeditempurchasecount * groupeditemavgprice
groupeditemtotalpurchasevalue

groupeditem_df = pd.DataFrame(columns = ['Purchase Count', 'Item Price', 'Total Purchase Value'])

groupeditem_df['Purchase Count'] = groupeditempurchasecount
groupeditem_df['Item Price'] = groupeditemavgprice
groupeditem_df['Total Purchase Value'] = groupeditemtotalpurchasevalue
groupeditem_df = groupeditem_df.sort_values('Purchase Count', ascending = False)

groupeditem_df.head()
```
